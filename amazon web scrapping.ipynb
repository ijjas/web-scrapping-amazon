```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome('chromedriver.exe')
```

```python
url='https://www.amazon.in/'
driver.get(url)
```

```python
search_box=driver.find_element(By.ID,"twotabsearchtextbox")
search_button=driver.find_element(By.XPATH,"//*[@id='nav-search-submit-button']")
```

```python
search_box.send_keys("smartphones under 10,000")
sleep(3)
search_button.click()
```

```python
products=[]
for i in range(10):
    print('scraping page',i+1)
    product=driver.find_elements(By.XPATH,"//span[@class='a-size-medium a-color-base a-text-normal']")
    for p in product:
        products.append(p.text)
    next_button=driver.find_elements(By.XPATH,"//*[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    sleep(2)
  
```

```python
len(products)
```

```python
products[:]
```

```python
reviews=[]
for i in range(10):
    print('scraping page',i+1)
    price=driver.find_elements(By.XPATH,"//span[@class='a-size-base s-underline-text']")
    for p in price:
        reviews.append(p.text)
    next_button=driver.find_elements(By.XPATH,"//*[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    sleep(2)
```

```python
reviews[:]
```

```python
len(reviews)
```

```python
deliveries=[]
for i in range(10):
    print('scraping page',i+1)
    delivery=driver.find_elements(By.XPATH,"//span[@class='a-color-base a-text-bold']")
    for p in delivery:
        deliveries.append(p.text)
    next_button=driver.find_elements(By.XPATH,"//*[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    sleep(2)
```
